```python
import pandas as pd

# Step 1: Load Data
df_logs = pd.read_csv(r'C:\Users\Cecilia\Downloads\ironhack\coursework\group_work\group_project_week5_6\second_project\data\clean\df_merged.csv')
df_visitids = pd.read_csv(r'C:\Users\Cecilia\Downloads\ironhack\coursework\group_work\group_project_week5_6\second_project\data\clean\merged_data.csv')

df_visitids.columns
```

```python
# Step 2: Merge the DataFrames on 'client_id' using an inner join
merged_df = pd.merge(df_logs, df_visitids, on='client_id', how='inner')

# Step 3: Sort the merged DataFrame by 'client_id' to keep it organized
merged_df = merged_df.sort_values(by='client_id').reset_index(drop=True)

# Display the first few rows of the merged DataFrame
merged_df.head()
```
